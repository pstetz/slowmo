```python
### Read the raw data
hc  = pd.read_csv(join(data_dir, "redcap_hc.csv"), low_memory=False)
mdd = pd.read_csv(join(data_dir, "redcap_mdd.csv"), low_memory=False)

### Remove unneeded rows
mdd = mdd[mdd.phq9.notnull()]
hc  = hc[hc.phq9.notnull()]
mdd = mdd[mdd.participant_id != "999_VIP_dontuse"]
hc  = hc[hc.participant_id   != "VIP001"]

### Format the subject id
mdd["subject"] = mdd.participant_id.map(lambda x: "conn%s" % (x.replace(" (s)", "")))
hc["subject"]  = hc.participant_id.map(lambda x: "conn%s" % x.rjust(3, "0"))

# mdd = mdd[["subject", "phq9"]]
# hc  = hc[["subject", "phq9"]]
# redcap = pd.concat([hc, mdd])
```
